Checkpoint Objective: Using the python data structures, create a program that allows the user to manage their shopping list. The user should be able to add items to the list, remove items, and view the current list of items.

STEPS:
1. Create a list named 'shopping_list' to store the items.

2. Use a while loop to create a menu of options for the user to add, remove, or view items from the list.

3. Use the input() function to prompt the user to make a selection from the menu.

4. Use an if-elif-else block to determine the user's selection and perform the corresponding action.

5. If the user selects 'add', use the input() function to prompt the user to enter an item to add to the list.

6. Use the range() function to limit the number of items that can be added to the list.

7. If the user selects 'remove', use the input() function to prompt the user to enter an item to remove from the list.

8. If the user selects 'view', use a for loop to iterate through the list of items and display them to the user.

9. Use the list, tuple, set, and dictionary data structures to store and manipulate the shopping items.


```python
#DATA STRUCTURE : LIST
shopping_list = ["rice"]
max_items = 2 # Define the maximum number of items allowed in the list

while shopping_list:
    selection = input("menu: select 'ADD', 'REMOVE' or 'VIEW'\n")
    if selection == "ADD":
        if len(shopping_list) <= max_items:
            item = input("Add an item to this list\n")
            shopping_list.append(item)
            print(shopping_list)
        else:
            print(f"Sorry, the list is full. You can't add more than {max_items} items.")
    elif selection == "REMOVE":
        if len(shopping_list) >= 0 :
            item_to_remove = input("Select item to be remove: \n")
            if item_to_remove in shopping_list:
                shopping_list.remove(item_to_remove)
                print(f"Shopping list : {shopping_list}")
            else:
                print(f"{item_to_remove} is not in the list.")
        else:
            print("The list is empty. Nothing to remove.")
    elif selection == "VIEW": # View list
        if len(shopping_list) > 0:
            print("Shopping List:")
            for item in shopping_list:
                print(f"- {item}")
        else:
            print("The list is empty.")
    else:
        print("Invalid choice. Please choose a valid option ('ADD', 'REMOVE', 'VIEW').")
    continue
        # End of the program


#From the code above, i have two observations:
#1. if i declare shopping_list as an empty list, the code will not run
#2. the while loop terminates, after all the items in the list have been removed (empty list)
```
